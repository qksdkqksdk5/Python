```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정 (필수!)
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우
plt.rcParams['axes.unicode_minus'] = False

# 예제 데이터
df = pd.DataFrame({
    '날짜': ['2025-01-01']*4 + ['2025-01-02']*4,
    '지역': ['서울','부산','대구','인천']*2,
    '상품': ['A','B']*4,
    '매출': [100,150,80,120, 130,170,90,140]
})
```

```python
# 1. 기본 막대그래프 (가장 많이 씀)
```

```python
df.pivot_table(index='날짜', columns='지역', values='매출', aggfunc='sum') \
  .plot(kind='bar', stacked=False, figsize=(10,6), title='지역별 매출')
plt.show()
```

```python
# 2. 누적 막대그래프 (전체 점유율 보기 좋음)
```

```python
df.pivot_table(index='날짜', columns='지역', values='매출') \
  .plot(kind='bar', stacked=True, figsize=(10,6), title='지역별 매출 (누적)')
plt.show()
```

```python
# 3. 선 그래프 (시간 흐름에 최고)
```

```python
df.pivot_table(index='날짜', columns='지역', values='매출') \
  .plot(kind='line', marker='o', figsize=(12,6), title='지역별 매출 추이')
plt.show()
```

```python
# 4. Seaborn으로 더 예쁘게 (추천!)
```

```python
pivot = df.pivot_table(index='날짜', columns='지역', values='매출', aggfunc='sum')

sns.heatmap(pivot, annot=True, fmt=',.0f', cmap='YlOrRd', linewidths=1)
plt.title('지역별 매출 히트맵')
plt.show()
```
