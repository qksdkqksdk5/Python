```python
pivot=pd.pivot_table(df, index='기준시', columns='특수일구분', values='교통량(대)')
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'   # 한글 폰트 설정 (Windows 기준 '맑은 고딕')

x=pivot.index

plt.plot(x, pivot['설날'], label='설날')
plt.plot(x, pivot['추석'], label='추석')
plt.plot(x, pivot['연말연시'], label='연말연시')
plt.legend()
plt.show()
```

```python
x=pivot.index  # x축 값: 피벗테이블의 index (예: 기준시 0~23)

plt.figure(figsize=(12,6))
plt.bar(x - 0.3,pivot['설날'], label='설날', width=0.3)
plt.bar(x,pivot['추석'], label='추석', width=0.3)
plt.bar(x + 0.3,pivot['연말연시'], label='연말연시', width=0.3)

plt.xticks(x, x)     # x축 눈금 위치와 표시값 설정 (index 그대로 표시)
plt.legend()         # 범례 표시
plt.tight_layout()   # 레이아웃 자동 조정 (레이블 겹침 방지)
plt.xlabel("기준시")
plt.ylabel("값")
plt.show()
```
